# Quadratic Forms - Loss Surfaces Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Quadratic Form Function
def quadratic_form(A, x):
    return x.T @ A @ x


# Matrix A (Positive Definite)
A = np.array([[3, 1], [1, 2]])


# Grid for visualization
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.zeros_like(X)


# Evaluate quadratic form for each point
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        x_vec = np.array([X[i, j], Y[i, j]])
        Z[i, j] = quadratic_form(A, x_vec)
```

## Contour Plot


```python
plt.figure(figsize=(6,5))
contours = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)
plt.title("Contour Plot of Quadratic Form")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

## 3D Surface Plot

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9)
ax.set_title("3D Surface of Quadratic Form")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x)")
plt.show()
```

## Gradient Descent on Quadratic Form

```python
def gradient(A, x):
    return 2 * A @ x


# Parameters
x = np.array([2.5, 2.5])
learning_rate = 0.1
iterations = 15
points = [x]


for i in range(iterations):
    grad = gradient(A, x)
    x = x - learning_rate * grad
    points.append(x)


points = np.array(points)


# Plot Gradient Descent on Contour
plt.figure(figsize=(6,5))
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.plot(points[:,0], points[:,1], 'ro--', label='Gradient Descent Path')
plt.title("Gradient Descent on Quadratic Form")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```
