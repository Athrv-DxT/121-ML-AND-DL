## Bias Variance Tradeoff

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Simulated data

```python
complexity = np.linspace(1, 10, 100)
bias2 = np.exp(-0.4 * complexity) * 0.5
variance = np.exp(0.4 * complexity - 4)
total_error = bias2 + variance + 0.1  # small irreducible error
```

## Graph

```python
plt.figure(figsize=(8, 5))
plt.plot(complexity, bias2, label="Bias²", linewidth=2)
plt.plot(complexity, variance, label="Variance", linewidth=2)
plt.plot(complexity, total_error, label="Total Error", linewidth=2, linestyle="--")
plt.xlabel("Model Complexity")
plt.ylabel("Error")
plt.title("Bias-Variance Tradeoff")
plt.legend()
plt.grid(True)
plt.show()
```
