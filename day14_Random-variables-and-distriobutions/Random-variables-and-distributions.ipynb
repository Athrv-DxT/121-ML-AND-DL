# Random Variables and Distributions :-

```python
!pip install scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, norm, uniform
```

## Discrete Random Variables

```python
# Bernoulli Distribution

p = 0.5
bernoulli_rv = bernoulli(p)
x_vals = [0, 1]
pmf_vals = bernoulli_rv.pmf(x_vals)

plt.bar(x_vals, pmf_vals)
plt.title("Bernoulli Distribution (p=0.5)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()
```

```python
# Binomial Distribution (n=10, p=0.5)

n, p = 10, 0.5
binom_rv = binom(n, p)
x = np.arange(0, n+1)
pmf = binom_rv.pmf(x)

plt.bar(x, pmf)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()
```

## Continous Random Variables 

```python
# Normal Distribution

mu, sigma = 0, 1
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.title("Normal Distribution (mu=0, sigma=1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Uniform Distribution

a, b = 0, 1
x = np.linspace(a, b, 100)
pdf_uniform = uniform.pdf(x, a, b-a)

plt.plot(x, pdf_uniform)
plt.title("Uniform Distribution [0,1]")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Expectation & Variance 

# Using Binomial distribution
E_X = binom_rv.mean()
Var_X = binom_rv.var()

print("Expectation E[X]:", E_X)
print("Variance Var[X]:", Var_X)
```
