# Matrix Calculus

### - Gradient of scalar function f(x): ∇f(x) = [∂f/∂x1, ∂f/∂x2, ..., ∂f/∂xn]^T
### - Jacobian of vector function f(x): J = [∂fi/∂xj]  (Matrix of partial derivatives)
### - Hessian (optional): Second-order derivatives matrix


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Gradient for scalar functions


```python
def scalar_function(x, y):
    return x**2 + y**2

def gradient_scalar(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

x, y = 3.0, 4.0
grad = gradient_scalar(x, y)
print("Gradient of f(x,y)=x²+y² at (3,4):", grad)
print()
```

## Jacobian for vector-valued functions

```python
def vector_function(x, y):
    f1 = x**2 + y
    f2 = x * y
    return np.array([f1, f2])

def jacobian_vector(x, y):
    df1_dx = 2*x
    df1_dy = 1
    df2_dx = y
    df2_dy = x
    return np.array([[df1_dx, df1_dy],
                     [df2_dx, df2_dy]])

J = jacobian_vector(x, y)
print("Jacobian at (3,4):", J)
```

## Gradient Descent Visualization for f(x)=x²

```python
f = lambda x: x**2
df = lambda x: 2*x

x_vals = [8]  
alpha = 0.1   
for i in range(10):
    x_new = x_vals[-1] - alpha * df(x_vals[-1])
    x_vals.append(x_new)

X = np.linspace(-9, 9, 100)
Y = f(X)
plt.figure(figsize=(7,5))
plt.plot(X, Y, label="f(x)=x²")
plt.scatter(x_vals, [f(x) for x in x_vals], color='red', label="Descent Path")
plt.title("Gradient Descent on f(x)=x²")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

## Applications in ML: Linear Regression Gradient

```python
np.random.seed(0)
X = 2 * np.random.rand(100,1)
y = 4 + 3*X + np.random.randn(100,1)

X_b = np.c_[np.ones((100,1)), X]

w = np.random.randn(2,1)

def compute_gradient(X_b, y, w):
    m = len(y)
    return -(2/m) * X_b.T @ (y - X_b @ w)

grad_w = compute_gradient(X_b, y, w)
print("Gradient of MSE Loss at initial weights:", grad_w)
```
