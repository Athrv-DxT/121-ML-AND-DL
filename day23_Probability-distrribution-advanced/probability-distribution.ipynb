# Advanced Probability Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2, t

sns.set(style="whitegrid")
```

## Chi-Square Distribution Simulation


```python
df_values = [2, 5, 10]  # degrees of freedom
x_chi = np.linspace(0, 30, 500)

plt.figure(figsize=(12,5))
for df in df_values:
    plt.plot(x_chi, chi2.pdf(x_chi, df), label=f"df={df}")
plt.title("Chi-Square Distribution PDFs")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## t-Distribution Simulation

```python
df_t = [1, 5, 10, 30]  # degrees of freedom
x_t = np.linspace(-5, 5, 500)

plt.figure(figsize=(12,5))
for df in df_t:
    plt.plot(x_t, t.pdf(x_t, df), label=f"df={df}")
plt.plot(x_t, 1/np.sqrt(2*np.pi)*np.exp(-x_t**2/2), 'k--', label="Standard Normal")
plt.title("t-Distribution vs Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## Random Sampling Examples

```python
# Chi-square sampling
sample_chi = chi2.rvs(df=5, size=1000)

# t-distribution sampling
sample_t = t.rvs(df=10, size=1000)

# Histograms
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(sample_chi, kde=True, color="b")
plt.title("Chi-Square Samples (df=5)")

plt.subplot(1,2,2)
sns.histplot(sample_t, kde=True, color="g")
plt.title("t-Distribution Samples (df=10)")
plt.tight_layout()
plt.show()
```
