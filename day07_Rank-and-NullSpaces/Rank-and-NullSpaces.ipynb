# Rank And Null Spaces

```python
import numpy as np
from numpy.linalg import matrix_rank
from scipy.linalg import null_space
import matplotlib.pyplot as plt
```

```python
A = np.array([[2, 4],
              [1, 2]])
print("Matrix A:\n", A)
print()

#Rank of the matrix
rankA = matrix_rank(A)
print("Rank of A:", rankA)
print()

#Null space of the matrix
null_A = null_space(A)
print("Null Space of A:\n", null_A)
print()

# Verify A * Null = 0
print("Verification A * Null = \n", A @ null_A)
```

## visualization

```python
if A.shape[1] == 2:
    origin = np.zeros((2,))
    plt.quiver(*origin, null_A[0, :], null_A[1, :], angles='xy', scale_units='xy', scale=1, color=['r'], label='Null Space Vector')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axhline(0, color='k')
    plt.axvline(0, color='k')
    plt.legend()
    plt.title("Null Space Vector Visualization")
    plt.grid()
    plt.show()
    
```
