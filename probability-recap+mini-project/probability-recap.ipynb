# Probability Recap

```python
import numpy as np
import matplotlib.pyplot as plt

# Basic Probability 
events = np.random.choice(['H', 'T'], size=1000, p=[0.5, 0.5])
p_H = np.mean(events == 'H')
p_T = np.mean(events == 'T')
print(f"P(H) = {p_H}, P(T) = {p_T}")

# Conditional Probability
data = np.random.choice(['A', 'B'], size=1000, p=[0.6, 0.4])
labels = np.random.choice(['Yes', 'No'], size=1000, p=[0.7, 0.3])
p_yes_given_A = np.mean(labels[data == 'A'] == 'Yes')
print(f"P(Yes | A) = {p_yes_given_A}")

# Bayes Theorem Example
# P(A|B) = (P(B|A) * P(A)) / P(B)
p_A = 0.3
p_B_given_A = 0.8
p_B = 0.5
p_A_given_B = (p_B_given_A * p_A) / p_B
print(f"P(A|B) = {p_A_given_B}")

# Law of Large Numbers
sample_sizes = np.arange(10, 1000, 50)
means = [np.mean(np.random.randn(n)) for n in sample_sizes]
plt.plot(sample_sizes, means, marker='o')
plt.axhline(0, color='r', linestyle='--')
plt.title("Law of Large Numbers")
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.show()

# Central Limit Theorem
n_experiments = 1000
sample_means = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(n_experiments)]
plt.hist(sample_means, bins=30, density=True, alpha=0.7)
plt.title("Central Limit Theorem - Sample Means")
plt.show()
```
