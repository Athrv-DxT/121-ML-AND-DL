# Gram-Schmidt Process and Orthonormal 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = np.array([[1, 1],
              [1, -1]])


def gram_schmidt(V):
    """
    Apply Gram-Schmidt process to get orthonormal basis.
    
    """
    n, m = V.shape
    Q = np.zeros((n, m))
    
    for i in range(m):
        # Take the current column vector
        q = V[:, i]
        
        # Subtract projections on previous basis vectors
        for j in range(i):
            q = q - np.dot(Q[:, j], V[:, i]) * Q[:, j]
        
        # Normalize to make it orthonormal
        Q[:, i] = q / np.linalg.norm(q)
    
    return Q

Q = gram_schmidt(V)


print("Original Matrix V:\n", V)
print()

print("Orthonormal Basis Q:\n", Q)
print()

I_check = np.round(np.dot(Q.T, Q), 5)
print("Q^T Q (should be Identity):\n", I_check)   #verification

```

## Visualization 


```python
origin_x = [0, 0]  
origin_y = [0, 0]  

plt.figure(figsize=(6, 6))
plt.quiver(origin_x, origin_y, V[0, :], V[1, :], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vectors')

plt.quiver(origin_x, origin_y, Q[0, :], Q[1, :], angles='xy', scale_units='xy', scale=1, color='r', label='Orthonormal Basis')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid()
plt.legend()
plt.title("Gram-Schmidt Orthonormalization")
plt.show()
```
