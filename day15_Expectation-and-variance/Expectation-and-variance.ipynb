# Expectation and Variance

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Expectation (Mean) & Variance 

```python
data = np.array([2, 4, 6, 8, 10])

# Mean (Expectation)

mean_val = np.mean(data)
print("Mean (Expectation):", mean_val)
print()

# Population Variance (dividing by n)
var_population = np.var(data)
print("Population Variance:", var_population)
print()

# Sample Variance (dividing by n-1)
var_sample = np.var(data, ddof=1) # ddof = Delta Degrees of Freedom
print("Sample Variance:", var_sample)
print()

# Standard Deviation
std_dev = np.std(data)
print("Standard Deviation:", std_dev)
print()
```

## Variance using Expectation Formula


```python
#Var(X) = E[X**2] - (E[X])**2

E_X = np.mean(data)
E_X2 = np.mean(data**2)
variance_formula = E_X2 - (E_X)**2
print("Variance using E[X**2] - (E[X])**2:", variance_formula)
```

## Law of Large Numbers Demonstration

```python
np.random.seed(42)
sample_sizes = np.arange(10, 1000, 10)
sample_means = [np.mean(np.random.randn(size)) for size in sample_sizes]

plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, sample_means, label="Sample Mean")
plt.axhline(y=0, color='r', linestyle='--', label="True Mean (0)")
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Law of Large Numbers")
plt.legend()
plt.show()
```

## Discrete Random Variable Expectation

```python
values = np.array([1, 2, 3, 4])
probs = np.array([0.1, 0.3, 0.4, 0.2])
E_discrete = np.sum(values * probs)
print("Discrete Expectation:", E_discrete)
```

## Continuous Random Variable Expectation 

```python
x = np.linspace(0, 1, 1000)
pdf = 2*x  # PDF: f(x) = 2x for x in [0,1]
E_continuous = np.trapezoid(x * pdf, x)  # Approx integral using trapz
print("Continuous Expectation:", E_continuous)
```
