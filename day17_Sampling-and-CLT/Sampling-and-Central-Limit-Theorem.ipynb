# Sampling & Central Limit Theorem (CLT)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Random Sampling Basics

```python
# Population: Uniform distribution

population = np.random.uniform(low=0, high=100, size=100000)

# Draw random samples

sample1 = np.random.choice(population, size=50, replace=False)
sample2 = np.random.choice(population, size=200, replace=False)

print("Sample 1 Mean:", np.mean(sample1))
print()
print("Sample 2 Mean:", np.mean(sample2))
```

##  Sampling Distribution of Mean

```python
sample_sizes = [10, 30, 50, 100]
means_dict = {}

for n in sample_sizes:
    sample_means = [np.mean(np.random.choice(population, size=n, replace=True)) for _ in range(1000)]
    means_dict[n] = sample_means

# Plotting

plt.figure(figsize=(10, 6))
for n, means in means_dict.items():
    sns.kdeplot(means, label=f"n={n}", fill=True)
plt.axvline(np.mean(population), color='r', linestyle='--', label="True Mean")
plt.title("Sampling Distribution of Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## Central Limit Theorem Demonstration

```python
# CLT: Distribution of sample means ~ Normal as n -> infinity

sample_means_clt = [np.mean(np.random.choice(population, size=50, replace=True)) for _ in range(1000)]

plt.figure(figsize=(8,5))
sns.histplot(sample_means_clt, bins=30, kde=True, color='g')
plt.title("Central Limit Theorem: Sample Means ~ Normal Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

## Law of Large Numbers vs CLT

```python
# LLN: Sample mean → True mean as n -> infinity

sample_sizes = np.arange(10, 1000, 10)
sample_means = [np.mean(np.random.choice(population, size=n, replace=True)) for n in sample_sizes]

plt.figure(figsize=(8,5))
plt.plot(sample_sizes, sample_means, label="Sample Mean")
plt.axhline(np.mean(population), color='r', linestyle='--', label="True Mean")
plt.title("Law of Large Numbers: Convergence of Sample Mean")
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.legend()
plt.show()
```
